## 問題7: 販売した商品の仕入先ごとの①販売数量の合計と②売上金額の合計が最も高い仕入先はどれか？

> （ヒント: pd.merge(sales_df,products_df[["商品ID","仕入先"]])）

```python
df = pd.merge(sales_df,products_df[["商品ID","仕入先"]])
df.head()
```

```python
df1 = df.groupby("仕入先").agg({
    "販売数量": "sum",
    "売上金額": "sum"
}).rename(columns={
    "販売数量":"総販売数量",
    "売上金額":"総売上金額"
})
df1
```
